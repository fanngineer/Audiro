```python
years=[]
for year in range(1990,2023):
    years.append(year)
```

```python
pops=[]
for year in years:
    driver.get(f'https://www.melon.com/chart/age/index.htm?chartType=YE&chartGenre=KPOP&chartDate={year}')
    sleep(0.1)
    tbody=driver.find_element_by_xpath('//*[@id="frm"]/table/tbody')
    trs=tbody.find_elements_by_tag_name('tr')
    for i in range(50):
        try:
            tds = trs[i].find_elements_by_tag_name('td')
            song_img = song_img = tds[2].find_element_by_tag_name('img').get_attribute('src').replace('/melon/resize/120/quality/80/optimize', '')
            song_title = tds[2].find_element_by_xpath(f'/html/body/div/div[3]/div/div/div[4]/div/div[2]/div[1]/form/table/tbody/tr[{i+1}]/td[4]/div/div/div[1]/span/strong/a').get_attribute('title').replace('#','')
            singer = tds[2].find_elements_by_xpath(f'/html/body/div/div[3]/div/div/div[4]/div/div[2]/div[1]/form/table/tbody/tr[{i+1}]/td[4]/div/div/div[2]/div[1]/a')[0].text.replace('#','')
            print(f' : {singer} - {song_title}')
            obj={'SONG_IMG':song_img,'SONG_TITLE':song_title,'SINGER':singer}
            pops.append(obj)
        except:
            print(i)
    print(f'한국{year}')
    driver.get(f'https://www.melon.com/chart/age/index.htm?chartType=YE&chartGenre=POP&chartDate={year}')
    sleep(0.1)
    tbody=driver.find_element_by_xpath('//*[@id="frm"]/table/tbody')
    trs=tbody.find_elements_by_tag_name('tr')
    for i in range(50):
        try:
            tds = trs[i].find_elements_by_tag_name('td')
            song_img = song_img = tds[2].find_element_by_tag_name('img').get_attribute('src').replace('/melon/resize/120/quality/80/optimize', '')
            song_title = tds[2].find_element_by_xpath(f'/html/body/div/div[3]/div/div/div[4]/div/div[2]/div[1]/form/table/tbody/tr[{i+1}]/td[4]/div/div/div[1]/span/strong/a').get_attribute('title').replace('#','')
            singer = tds[2].find_elements_by_xpath(f'/html/body/div/div[3]/div/div/div[4]/div/div[2]/div[1]/form/table/tbody/tr[{i+1}]/td[4]/div/div/div[2]/div[1]/a')[0].text.replace('#','')
            print(f' : {singer} - {song_title}')
            obj={'SONG_IMG':song_img,'SONG_TITLE':song_title,'SINGER':singer}
            pops.append(obj)
        
        except:
            print(i)
    print(f'미국{year}')
    
        
print(pops)
```

```python
def get_song_url(song_title, singer):
    keyword = song_title.strip().replace(' ', '+') + singer.strip().replace(' ', '+')
    driver.get(f'https://www.youtube.com/results?search_query={keyword}')
    sleep(0.5)
    video_ids=driver.find_elements_by_xpath('//*[@id="thumbnail"]')
    print(len(video_ids))
    for id in video_ids:
        try:
            id.click()
            sleep(0.5)
            
        except:
            pass
    song_url=driver.current_url
    return song_url
```

```python
for song in pops:
    song_title=song['SONG_TITLE']
    singer=song['SINGER']
    try:
        song['SONG_URL']= get_song_url(song_title,singer)
    except:
        print(song)
song

        
```
